```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x= np.array([2100, 1600, 2400, 1400, 3000])
y=np.array([400, 330, 369, 290, 540])
```

```python
def cost_function(x,y,w,b):
    m=len(x)
    cost=0
    for i in range(len(x)):
        pre=(w * x[i]) + b
        cost+=(pre-y[i])**2
    return cost / (2*m)
```

```python
def gradient(x,y,w,b):
    m=len(x)
    dw=0
    db=0
    for i in range(m):
        pre=(w * x[i]) + b
        dw+=(pre-y[i])*x[i]
        db+=(pre-y[i])
    return dw/m,db/m
```

```python
def gradientdecent(w,b,learning,x,y,iteration):
   
    for i in range(iteration):
        dw,db=gradient(x,y,w,b)
        w-=learning*dw
        b-=learning*db
        # cost=cost_function(x,y,w,b)
    return w,b
```

```python
w_init = 0
b_init = 0
alpha = 0.0000001 


iterations = 1000

w_final, b_final = gradientdecent(w_init,b_init,alpha,x,y,iterations)

print("\nTrained parameters: w = ",w_final,"b = ",b_final)
```

```python
y_pred = w_final * x + b_final
plt.scatter(x, y, label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('Size (sqft)')
plt.ylabel('Price ($1000)')
plt.title('Linear Regression with scikit-learn')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([2100, 1600, 2400, 1400, 3000]).reshape(-1, 1)
y = np.array([400, 330, 369, 290, 540])

model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)

print("Weight (w):", model.coef_[0])
print("Bias (b):", model.intercept_)
print(y)
plt.scatter(x, y, label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('Size (sqft)')
plt.ylabel('Price ($1000)')
plt.title('Linear Regression with scikit-learn')
plt.legend()
plt.show()
```
